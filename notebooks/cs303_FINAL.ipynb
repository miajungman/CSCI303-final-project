```python
import pandas as pd
import matplotlib.pyplot as plt

# Load game.csv into a DataFrame
game_df = pd.read_csv('game.csv')

# Load imdb_video_game_rating.csv into a DataFrame
imdb_df = pd.read_csv('imdb_video_game_rating.csv')

# Merge the two DataFrames based on their common column 'video_game' and 'title'
merged_df = pd.merge(game_df, imdb_df, left_on='video_game', right_on='title')

# Drop the redundant 'title' column
merged_df.drop('title', axis=1, inplace=True)

# Create a scatter plot with different color for game.csv 'rank' vs mdb_video_game_rating.csv 'rating'
plt.scatter(merged_df['video_game'], merged_df['Rank'], c=merged_df['rating'])
plt.xticks(rotation=90)
plt.xlabel('Video Game')
plt.ylabel('Rank')
plt.title('Video Game Rank vs IMDB Rating')
plt.colorbar()
plt.show()
```
